Указываю пути к моделям , а так как они лежат в одной папке с проектом , то просто указываю имена

```python
predictor_path = 'shape_predictor_5_face_landmarks.dat'
face_rec_model_path = 'dlib_face_recognition_resnet_model_v1.dat'
```

Импортирую модели распознавания

```python
detector = dlib.get_frontal_face_detector()
sp = dlib.shape_predictor(predictor_path)
facerec = dlib.face_recognition_model_v1(face_rec_model_path)
```

Загружаю фотографию и получаю , сначало масив всех обнаруженых лиц на фото (dets)
получаю именно точки где находится лицо человека(d)
получаю ландмарки (shape)
получаю дескрипторы(face_descriptor)

```python



img = cv2.imread('putin_1.jpg')
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
dets = detector(rgb_image, 1)
for k, d in enumerate(dets):
    print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
        k, d.left(), d.top(), d.right(), d.bottom()))

    
    
shape = sp(rgb_image, d)
face_descriptor = facerec.compute_face_descriptor(rgb_image, shape)
```
